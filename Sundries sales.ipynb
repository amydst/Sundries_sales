```python
import pandas as pd
import os
import re
```

```python

item_sales_df = pd.read_excel('Monthly sales\February 2025.xlsx')

item_sales_df.head(5)
```

```python
folder = 'Monthly sales'

months_list= []

# List all files in the folder
for file in os.listdir(folder):
    file_path = os.path.join(folder, file)  # Get full file path
    
    if file.endswith('.xlsx'):  # Process only Excel files
        months_df = pd.read_excel(file_path)  # Read the file using full path
        months_df['Month'] = re.sub(r".{5}$", "", file) 
        months_list.append(months_df)
```

```python
# Concatenate all DataFrames in the list into a single DataFrame
item_sales_df = pd.concat(months_list, ignore_index=True)

```

```python
# Display the first few rows of the concatenated DataFrame 
item_sales_df.head(5)
        
```

```python
folder_costs = '3612'

dept_list= []

# List all files in the folder
for files in os.listdir(folder_costs):
    file_path = os.path.join(folder_costs, files)  # Get full file path
    print(file_path)
    
    if file.endswith('.xlsx'):  # Process only Excel files
        dept_df = pd.read_excel(file_path)  # Read the file using full path
        dept_list.append(dept_df)

cost_df = pd.concat(dept_list, ignore_index=True)

# Display the first few rows of the concatenated DataFrame
cost_df.head(5)

```

```python
cost_df = cost_df[cost_df['Retail'].notna()]


```

```python
merged_item_sales_df = pd.merge(item_sales_df, cost_df, on=['UPC'], how='left')


# Display the first few rows of the merged DataFrame
merged_item_sales_df.head(5)

```
